```python
#Introducing a Deck class
import random
class Deck():
    number_of_cards = 52
    def __init__(self):
        self.deck = []
        for suit in Card.suits:
            for rank in Card.ranks:
                card = Card(suit,rank)
                self.deck.append(card)
    def __str__(self):
        self.deck2 = []
        for card in self.deck:
            self.deck2.append(str(card))
        return self.deck2
    def get_deck(self):
        return self.deck
    def shuffle_deck (self):
        random.shuffle(self.deck)
    def deal_card (self):
        dealing_card = self.deck[-1]
        self.deck.pop()
        self.number_of_cards = self.number_of_cards - 1
        return dealing_card
```
