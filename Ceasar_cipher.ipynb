```python
'''This is my implementation of deciphering Caesar cipher.
Assuming we don't know how many turns down the alphabet we need,
the program generates all 26 possible version of given text
leaving it for the user to decide which version makes any sense.
'''

alphabet = 'abcdefghijklmnopqrstuvwxyz'
alphabet_list = list (alphabet)
        
def message_to_number (message):
    encoded_message_numeric = []
    for letter in message:
        index = alphabet_list.index(letter)
        encoded_message_numeric.append(index)
    return encoded_message_numeric

def make_turn (encoded_message_numeric):
    for counter in range (1, len(encoded_message_numeric)+1):
        encoded_message_numeric [counter -1] = encoded_message_numeric [counter -1] + 1
        if encoded_message_numeric [counter -1] == 27:
            encoded_message_numeric [counter -1] = 1
            continue
    return encoded_message_numeric

def number_to_letter (encoded_message_numeric):
    encoded_message_letters = []
    for digit in encoded_message_numeric:
        encoded_message_letters.append(alphabet_list[digit-1])
    return encoded_message_letters

print ("""Welcome to decoding Caesar cipher!  
It is a type of substitution cipher 
in which each letter in the plaintext is 'shifted' 
a certain number of places down the alphabet.
""")

encoding_message = str (input('Please enter a message to decipher. '))

encoding_message = encoding_message.lower()
my_mes2=[]
for char in encoding_message:
    if char.isalpha():
        my_mes2.append(char)
    else:
        continue
encoding_message = ''.join(my_mes2)
print ('Romans used only lowercase letters and no spaces or numbers.')
print (encoding_message)
print ("We don't know how many turns down the alphabet are needed.")
print ('So I will provide you all possible versions one by one.')
print ('Please check if any of these makes any sense.\n')

encoding_message_numeric = message_to_number (encoding_message)

for n in range (1, 27):
    next_version = make_turn (encoding_message_numeric)
    print (''.join (number_to_letter(next_version)))
```
