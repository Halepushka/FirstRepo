```python
#Fibonacci Sequence
#Enter a number and have the program generate the Fibonacci sequence to that number or to the Nth number.
def fibonacci(n):
    a = 1
    b = 1
    while a <= n:
        yield a,b
        a,b = b, a+b
list = []
n = int(input ('Please set the number to generate Fibonacci sequence to that number '))
for items in fibonacci (n):
    list.append (items[0])
print (list)
```

```python
#Next Prime Number 
#Have the program find prime numbers until the user chooses to stop asking for the next one.

def get_primes(last_prime):
    for number in range (last_prime+1, 1000):
        for prime in range (2,number):
            if number%prime == 0:
                break
            else:
                continue
        else:
            yield number

last_prime = 2
answer = input('Do you want a prime number? y/n ')
if answer == 'y':
    print(last_prime)
            
while True:
    answer = input('Do you want a prime number? y/n ')
    if answer == 'y':
        try:
            current_prime = next(get_primes(last_prime))
            print (current_prime)
            last_prime = current_prime        
            continue
        except:
            print ('Limit reached')
            break
    else:
        break
```
